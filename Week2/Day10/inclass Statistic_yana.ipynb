```python
import pandas as pd

df = pd.DataFrame({'values': [10, 20, 30, 40, 50]})

variance = df['values'].var()  # default = sample variance (ddof=1)
population_variance = df['values'].var(ddof=0)

print("Sample Variance:", variance)
print("Population Variance:", population_variance)
```

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])
variance = np.var(data)          # population variance
sample_variance = np.var(data, ddof=1)  # sample variance

print("Variance:", variance)
print("Sample Variance:", sample_variance)
```

```python
import numpy as np

class1 = np.array([50, 47, 51, 49, 47])
class2 = np.array([5, 20, 50, 80, 130])

mean1 = np.mean(class1)
mean2 = np.mean(class2)

var1 = np.var(class1)
var2 = np.var(class2)

print("Class 1 Mean:", mean1)
print("Class 1 Variance:", var1)

print("Class 2 Mean:", mean2)
print("Class 2 Variance:", var2)
```

```python
import numpy as np

class1 = np.array([50, 47, 51, 49, 47])
class2 = np.array([5, 20, 50, 80, 130])

mean = sum(class1) / len(class1)
print("Mean of class1:", mean)

variance = ((class1 - mean) ** 2).sum() / len(class1)   # population variance
print("Variance of class1:", variance)

mean = sum(class2) / len(class2)
print("Mean of class2:", mean)

variance = ((class2 - mean) ** 2).sum() / len(class2)   # population variance
print("Variance of class1:", variance)

```

```python
std1 = variance ** 0.5
std2 = variance ** 0.5

print(std1, std2)
```
