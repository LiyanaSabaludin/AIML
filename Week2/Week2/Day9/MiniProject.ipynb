```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("Titanic-Dataset.csv")   
df
```

```python
#sex_counts.index → the labels (male, female)
#sex_counts.values → the numbers (counts)

sex_counts = df["Sex"].value_counts()
plt.bar(sex_counts.index, sex_counts.values, color='orange')

plt.title("Titanic Survivors based on Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Survivors")
plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sex_counts = df["Sex"].value_counts()
plt.bar(sex_counts.index, sex_counts.values, color='orange')
plt.title("Titanic passenger based on Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Survivors")
plt.show()

plt.subplot(1,3,2)
passenger_class = df["Pclass"].value_counts()
plt.bar(passenger_class.index, passenger_class.values, color='green')
plt.title("Passenger Count by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.show()

plt.subplot(1,3,3)
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers Survived")
plt.show()

```
