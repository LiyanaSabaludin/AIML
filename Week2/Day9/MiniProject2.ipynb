```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("Amazon.csv")
df
```

```python

plt.figure(figsize=(25,7))
payment = df["PaymentMethod"].value_counts()
plt.bar(payment.index, payment.values, color='orange')
plt.title('Payment Method Popularity in Amazon Users')
plt.xlabel('Payment Method')
plt.ylabel('Number of People')
plt.show()
```

```python
plt.figure(figsize=(20,7))
payment = df["PaymentMethod"].value_counts()
sns.scatterplot(x=payment.index, y=payment.values, color='orange')

plt.title("Number of Users by Payment Method")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example: if your data is in a CSV file
df = pd.read_csv("Amazon.csv")

# Now you can use pairplot
sns.pairplot(df, hue='PaymentMethod')
plt.show()

```

The bar chart shows how many people use each payment method,the pattern suggests people nowadays move toward cashless payments, with credit cards leading, digital methods growing, and cash becoming a least preferred
