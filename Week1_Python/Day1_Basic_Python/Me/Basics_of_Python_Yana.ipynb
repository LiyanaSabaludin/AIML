```python
#continue: skip that particular iteration and loop will continue after that
for i in range(1,100):
  if(i%5==0):
    continue
  print(i,end=' ')
```

```python
#assignment operator = e.g. username=sam1256
# equal == e.g if(number==5)
#= means assigning value, == means equal to
atm_pin=1243
allowed_attempts=3
while(allowed_attempts!=0):
  user_pin=int(input('Enter you pin:\t'))
  if(user_pin==atm_pin):
      print('Welcome to ATM')
      break
  else:
    allowed_attempts-=1

  if(allowed_attempts==0):
      print('Your card has been blocked')
  else:
      print('Wrong Pin Number, Try Again')
```

```python
hidden_number=7
allowed_attempts=3

print("Guess the hidden number between 1 and 10!")

while(allowed_attempts!=0):
    guess=int(input("Enter your guess: "))
    if(guess==hidden_number):
        print("Correct! You found the hidden number:")
        break
    else:
     allowed_attempts-=1

    if(allowed_attempts==0):
      print('Your Lost Your Chance')

    elif(guess<hidden_number):
        print("Too Low")

    elif(guess>hidden_number):
        print("Too High")

```
